```python
import numpy as np

import pandas as pd
from pandas import Series,DataFrame
from bokeh.layouts import gridplot, row, column

from bokeh.io import push_notebook, show, output_notebook, output_file
from bokeh.plotting import figure


# For very simple visualizations (without too much interaction, output_notebook works too, which we enable here)
output_notebook()
```

```python
df = pd.read_csv('../Data/clean/finalclean3.0.csv')
# om sjv te gebruiken, misschien sjv in bins gebruiken
data = DataFrame(df)
```

```python
# Welke variabele tegen elkaar worden uitgezet
X = np.vstack(data["%Slimme Meter"])
Y = data["%SJV laag tarief"]
```

```python
X = np.column_stack((X, np.ones(X.shape[0])))

# a en b berekenen
a, b = np.linalg.lstsq(X, Y)[0]
print(a, b)
```

```python
f = figure(plot_width=400, plot_height=250)
S = data["%Slimme Meter"]
# scatter-plot
f.circle(S, Y)
output_file('laagslim.html')
# Create the line
x = S
f.line(x, a * x + b, color='red')

## Add some axis information
f.xaxis.axis_label = '%Slimme Meter'
f.yaxis.axis_label = "SJV laag tarief"

show(f)
```
