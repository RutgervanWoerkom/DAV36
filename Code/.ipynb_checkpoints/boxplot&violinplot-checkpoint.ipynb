```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pandas.tools import plotting

# Belangrijk dat sjv daadwerkelijkj in bins staat!
data = pd.read_csv('../Data/clean/SJV_in_bin.csv')
sjv = data["SJV"]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

# # plot violin plot
parts = axes[0].violinplot(sjv,
                   showmeans=False,
                   showmedians=True)

for pc in parts['bodies']:
    pc.set_facecolor('lightgreen')
    pc.set_edgecolor('black')
    pc.set_alpha(1)
    

axes[0].set_title('violin plot')

# # plot box plot
axes[1].boxplot(sjv, showmeans=True, meanline=True)
axes[1].set_title('box plot')



plt.show()

```
